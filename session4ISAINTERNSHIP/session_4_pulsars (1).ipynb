### Binning of light curve
Bin size needs to be large enough for sufficient number of counts per bin but small enough so that real variation is not hidden

```python

def lc_binning(time, rate, binsize=2): #instead of 0.01 sec give data every 2 sec
    bins = np.arange(time[0], time[-1], binsize) #creates array from time[0] till time[-1] with stepsize of 2 i.e binsize
    digitized = np.digitize(time, bins)#here if time=[0,1,2,3,4] and bins=[0,2,4,6,8] then o/p will be [1,1,2,2,3] as 0 and 1 go in bin1,2 and3 go in bin2 and 4 goes in bin3

    abs_time_bin = np.array([time[digitized == i].mean() if len(time[digitized == i]) > 0 else np.nan for i in range(1, len(bins))]) #can find what data points come in which bins. the mean time in the every bin ig. if the number of times in each bin>0 then the mean of those times in the specific bin taken
    un_time_bin = abs_time_bin - abs_time_bin[0] #relative to the first time ig. as its taken as reference time

    un_rate_bin = np.array([rate[digitized == i].mean() if len(rate[digitized == i]) > 0 else np.nan for i in range(1, len(bins))]) #rate corresponding to the bin. same explanation as givem for time

    # Keep only indices where both time and rate are not NaN
    mask = ~np.isnan(un_time_bin) & ~np.isnan(un_rate_bin)  #nan values are removed

    time_bin = un_time_bin[mask] #applying the mask to the arrays
    rate_bin = un_rate_bin[mask] #applying the mask to the arrays

    return time_bin, rate_bin
```

```python
%matplotlib inline
time_bin,rate_bin=lc_binning(time,rate,2)#binning helps us look at the data clearly. if there is a line graph goes upward steap means more photons coming and pulsar facing towards us.if graph goes steep down then its fully away from us
plt.figure(figsize=(16,8))
plt.plot(time_bin,rate_bin,'-o')
plt.title('Light Curve')
plt.grid()
plt.xlabel('Time(s)')
plt.ylabel('Count/s')
```

Let's bin and fold the profile to recover pulsar

```python
def fold_light_curve(time, rate, period, nbins=32): #folding means the periodically we can see the power peak right? the values in all of these periodic peaks are added/mean taken. not only these values but all values are added/mean taken. its done so that the peak values stand out from the other values
    #nbins=32 means 32 values
    # Compute phase
    phase = (time % period) / period  # Normalize to [0, 1). From 0 to 1 only. till 4.76 sec period

    # Define bins
    bin_edges = np.linspace(0, 1, nbins + 1)  #from 0 to 1 with 33 spacing
    bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:]) #centres the bin
    folded_counts = np.zeros(nbins)

    # Bin the data
    indices = np.digitize(phase, bin_edges) - 1
    for i in range(nbins):
        bin_values = rate[indices == i]
        if bin_values.size > 0:
            folded_counts[i] = bin_values.mean() #.mean() can be .summation() also. for every bin the values are averaged

    # Repeat for two cycles
    two_cycle_phase = np.concatenate([bin_centers, bin_centers + 1])#from 0 to 2 it becomes. 0 to 1 is 4.76 second period. 1 to 2 is 4.76 second period
    two_cycle_counts = np.concatenate([folded_counts, folded_counts])#from 0 to 2 it becomes

    # Plotting
    plt.figure(figsize=(7, 4))
    plt.plot(two_cycle_phase, two_cycle_counts, drawstyle='steps-mid', lw=2)
    plt.xlabel('Phase (2 cycles)')
    plt.ylabel('Mean Rate')
    plt.title(f'Folded Light Curve (P = {period:.6f} s)')
    plt.grid(True)
    plt.xlim(0, 2)
    plt.show()
    return bin_centers, folded_counts
```

```python
# Assuming you have 'time' and 'rate' arrays
fold_light_curve(time, rate, period, nbins=32)
```
